```python
import pandas as pd
import openai
from tqdm import tqdm

client = openai.OpenAI()

# 读取数据
neighbor_data = pd.read_csv('./input/小区数据.csv', encoding='utf-8')
```

```python
# 创建一个空的 DataFrame 用于存储处理结果
processed_data = pd.DataFrame(columns=['小区信息-处理后'])

# 使用 tqdm 创建一个进度条
for index, row in tqdm(neighbor_data.iterrows(), total=len(neighbor_data), desc="处理进度"):
    # 获取当前行的 '小区信息' 的值
    original_data = row['小区信息']

    # 检查是否为缺失值，如果是，则跳过该行
    if pd.isna(original_data):
        # 如果为空值，将空值写入新的 CSV 文件
        processed_data.loc[index] = ""
        processed_data.to_csv('processed_data.csv', index=False)
        continue

    # 调用 OpenAI API 进行处理
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        temperature=0,
        max_tokens=100,
        messages=[
            {"role": "system",
             "content": "你是一个数据提取师，你需要提取一句话里的信息，你的回答格式必须为(没有的以None作为占位符)：绿化率:|建筑年代:|物业信息:|停车:"},
            {"role": "user", "content": f"{original_data}"},
        ]
    )

    # 获取处理后的数据
    processed_data_content = response.choices[0].message.content
    processed_data_content = "|".join(processed_data_content.split("\n"))

    # 将处理后的数据存放在 '小区信息-处理后' 列中
    processed_data.loc[index] = processed_data_content

    # 将结果保存到新的 CSV 文件中
    # processed_data.to_csv('./output/openai-neighbor.csv', index=False)
```
